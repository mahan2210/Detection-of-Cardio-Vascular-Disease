```python
details.groupby('Gender')['BMI'].value_counts()
```

```python
details.groupby('Gender')['BMI'].value_counts().unstack()
```

```python
details.groupby('Gender')['BMI'].value_counts().unstack().plot(kind='bar',stacked= True, figsize=(10,6))
```
